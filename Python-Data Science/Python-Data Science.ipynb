Importing data science libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading data from a csv file

```python
avoData=pd.read_csv('avocado.csv')
avoData
```

1. Displaying the number of unique values in each column.

```python

uniqueValues=avoData.nunique()
uniqueValues
```

2. Choosing any three columns, access them with bracket notation, and displaying the first five rows of this data.

```python

avoData[['Date','year','region']].head(5)
```

3. Selecting one column and accessing it with dot notation.

```python

avoData.region
```

4. Multiplying the Total Volume and AveragePrice columns and store the result in a new column
called EstimatedRevenue. Then, displaying the first five rows of this data to confirm that the
column was added and has the correct values.

```python
avoData['EstimatedRevenue']=avoData['Total Volume']*avoData['AveragePrice']
avoData
```

5.Creating a DataFrame that’s grouped by region and type and that includes the average price for
the grouped columns. Then, resetting the index and displaying the first five rows.

```python
grouped=avoData.groupby(['region','type'])['AveragePrice']
df_columns=['region&type','AveragePrice']
groupedDf=pd.DataFrame(data=grouped,columns=df_columns)
groupedDf.reset_index()
groupedDf.head(5)
```

6.Creating a Pandas plot to show the year in which the most bags of avocados were sold

```python
avoData['year'].value_counts().plot.bar()
```

7.Creating a bar plot that shows the mean, median, and standard deviation of the Total Volume
column by year.

```python
avoData.groupby('year')['Total Volume'].agg(['mean','median','std']).plot.bar()
```

8. Creating a scatterplot with a Seaborn specific method:

```python
sns.scatterplot(data=avoData,x='AveragePrice',y='Total Volume')
```

9. Creating the same plot with a generic method.

```python
sns.relplot(data=avoData,x='AveragePrice',y='Total Volume',kind='scatter')
```
